## topic 횟수 제한해보기

```python
import rclpy as rp
from turtlesim.msg import Pose
```

```python
rp.init()
test_node = rp.create_node('sub_test')
```

```python
cnt = 0
def callback(data):
    global cnt
    cnt += 1
    print(">", cnt, " -> X : ", data.x, ", Y : ", data.y)
    if cnt > 3:
        raise Exception("Subscription Stop")
```

## 다시 실행

```python
test_node.create_subscription(Pose, '/turtle1/pose', callback, 10)
```

## spin

```python
rp.spin(test_node)
```
